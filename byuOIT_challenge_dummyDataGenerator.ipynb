```python
from dateutil import parser
def dateGenerator(monStart, monEnd, yearStart, yearEnd):
  MonthDays = {30: [9,4,6,11], 31: [1,3,5,7,8,10,12], 28: [2]}
  year = random.randint(yearStart,yearEnd)
  month = random.randint(monStart, monEnd)
  day = 0
  date_string = ""
  if month in MonthDays[30]:
    day = random.randint(1,30)
  if month in MonthDays[31]:
    day = random.randint(1,31)
  if month in MonthDays[28]:
    day = random.randint(1,28)
  if month < 10:
    date_string = f"{year}-0{month}"
  else:
    date_string = f"{year}-{month}"
  if day < 10:
    date_string += f"-0{day}"
  else:
    date_string += f"-{day}"

  date = parser.parse(date_string).date()
  return date

def getAccessibilityAccomadations():
  accoms = ["This student requires extra time for test-taking due to a learning disability.", "This student benefits from having notes read aloud to them during lectures.", "This student needs frequent breaks in order to stay focused during class.", "This student struggles with organization and may benefit from additional support in that area.", "This student is visual learner and benefits from the use of visual aids in lessons."]
  if random.randint(0,99) > 84:
    return accoms[random.randint(0,4)]

import datetime
import random
def randomTimeGenerator():
  # Generate a random hour, minute, and second
  hour = random.randint(0, 23)
  minute = random.randint(0, 59)
  second = random.randint(0, 59)

  # Create a time object using the random values
  t = datetime.time(hour, minute, second)
  return t

def addressGenerator():
  NS = ["N", "S"]
  EW = ["E", "W"]
  house_num = random.randint(15, 10000)
  street_num = random.randint(15, 10000)
  return f"{house_num} {EW[random.randint(0,1)]} {street_num} {NS[random.randint(0,1)]}"
```

```python
import random
import uuid

class Student:
  def __init__(self, studentId, home_address_id, mailing_address_id):
    self.studentId = studentId
    self.first_name = random.choice(MF_first_names)
    self.last_name = random.choice(last_names)
    self.sex = random.choice(["M", "F"])
    self.dob = dateGenerator(1,12,1985,2009)
    self.home_address_id = home_address_id
    self.mailing_address_id = mailing_address_id
    self.phone_number = random.randint(1000000000,9999999999)
    self.personal_email = "student" + str(random.randint(1, 1000)) + "@example.com"
    self.work_email = "student" + str(random.randint(1, 1000)) + "@work.com"
    self.religion = random.choice(["LDS", "Non-LDS"]),
    self.ward = "Ward " + str(random.randint(1, 1000)),
    self.stake = "Stake " + str(random.randint(1, 1000)),
    self.accessibility_accommodations = getAccessibilityAccomadations()
    self.high_school = "high school " + str(random.randint(1,1000))
    self.high_school_gpa = round(random.uniform(2.75, 4.00))
    self.act_score = random.randint(19,36)
    self.sat_score = random.randint(400, 1600)
```

```python
class HomeAddress:
  def __init__(self, home_address_id):
    self.home_address_id = home_address_id
    self.street_address = addressGenerator()
    self.city = random.choice(cities)
    self.state_province = random.choice(states)
    self.country = "US"
    self.zip_code = str(random.randint(10000, 99999))

class MailingAddress:
  def __init__(self, mailing_address_id):
    self.mailing_address_id = mailing_address_id
    self.street_address = addressGenerator()
    self.city = random.choice(cities)
    self.state_province = random.choice(states)
    self.country = "US"
    self.zip_code = str(random.randint(10000, 99999))
    self.po_box = str(random.randint(10000, 99999))
```

```python
class Course:
  def __init__(self, course_id, instructor_id, course_name, course_title):
    self.course_id = course_id
    self.semester = random.choice(["Winter", "Fall", "Spring", "Summer"])
    self.course_name = course_name
    self.course_title = course_title
    self.instructor_id = instructor_id
    self.meeting_days = random.choice(["MW", "MWF", "TTH"])
    self.meeting_time = randomTimeGenerator()
    self.online_percentage = round(random.uniform(0.00, 1.00),2)
    self.min_credits = random.randint(1,4)
```

```python
class Instructor:
  def __init__(self, instructor_id):
    self.instructor_id = instructor_id
    self.first_name = random.choice(MF_first_names)
    self.last_name = random.choice(last_names)
```

```python
class Enrollment:
  def __init__(self, enrollment_id, student_id, course_id, min_credits):
    self.enrollment_id = enrollment_id
    self.student_id = student_id
    self.course_id = course_id
    self.credit_hours = min_credits
    self.enrollment_type = random.choice(["audit", "credit"])
```

```python
class AddDropWithdrawal:
  def __init__(self, add_drop_id, enrollment_id):
    self.id = add_drop_id
    self.enrollment_id = enrollment_id
    self.date = dateGenerator(1,12,2020,2023)
    self.done_by_student = random.choice([True, False])
    self.action_type = random.choice(["A", "D", "W"])
```

```python
class Grade:
  def __init__(self, grade_id, enrollment_id):
    self.grade_id = grade_id
    self.enrollment_id = enrollment_id
    self.grade = random.choice(["C-", "C", "C+", "B-", "B", "B+", "A-", "A"])
    self.gpa_equivalent = random.choice([1.7,2.0,2.3,2.7,3.0,3.4,3.7,4.0])
    self.counts_towards_cum_credits = random.choice([True, False])
    self.passing = random.choice([True, False])
```

```python
class Incomplete:
  def __init__(self, incomplete_id, enrollment_id):
    self.incomplete_id = incomplete_id
    self.enrollment_id = enrollment_id
    self.contract_deadline = dateGenerator(1,12,2020,2023)
    self.gpa_penalty_for_late = random.choice([1.7,2.0,2.3,2.7,3.0,3.4,3.7,4.0])
```

```python
students = []
homeAddresses = []
mailingAddresses = []
grades = []
incompletes = []
addDropsWithdrawals = []
enrollments = []
instructors = []
courses = []
for i in range(50):
  student = Student(i,i,i)
  students.append([student.studentId, student.first_name, student.last_name, student.sex, student.dob, student.home_address_id, student.mailing_address_id, student.phone_number, student.personal_email, student.work_email, student.religion, student.ward, student.stake, student.accessibility_accommodations, student.high_school, student.high_school_gpa, student.act_score, student.sat_score])

  home_address = HomeAddress(student.home_address_id)
  homeAddresses.append([home_address.home_address_id, home_address.street_address, home_address.city, home_address.state_province, home_address.country, home_address.zip_code])

  mailing_address = MailingAddress(student.mailing_address_id)
  mailingAddresses.append([mailing_address.mailing_address_id, mailing_address.street_address, mailing_address.city, mailing_address.state_province, mailing_address.country, mailing_address.zip_code, mailing_address.po_box])

  instructor = Instructor(i)
  instructors.append([instructor.instructor_id, instructor.first_name, instructor.last_name])

  course = Course(i, instructor.instructor_id, class_names[i % len(class_names)], class_titles[i % len(class_titles)])
  courses.append([course.course_id, course.semester, course.course_name, course.course_title, course.instructor_id, course.meeting_days, course.meeting_time, course.online_percentage, course.min_credits])

  enrollment = Enrollment(i, student.studentId, course.course_id, course.min_credits)
  enrollments.append([enrollment.enrollment_id, enrollment.student_id, enrollment.course_id, enrollment.credit_hours, enrollment.enrollment_type])

  grade = Grade(i, enrollment.enrollment_id)
  grades.append([grade.grade_id, grade.enrollment_id, grade.grade, grade.gpa_equivalent, grade.counts_towards_cum_credits, grade.passing])

  addDropWithdrawal = AddDropWithdrawal(i, enrollment.enrollment_id)
  addDropsWithdrawals.append([addDropWithdrawal.id, addDropWithdrawal.enrollment_id, addDropWithdrawal.date, addDropWithdrawal.done_by_student, addDropWithdrawal.action_type])

  incomplete = Incomplete(i, enrollment.enrollment_id)
  incompletes.append([incomplete.incomplete_id, incomplete.enrollment_id, incomplete.contract_deadline, incomplete.gpa_penalty_for_late])

```

```python
import pandas as pd

Student_df = pd.DataFrame(students, columns=["studentId", "first_name", "last_name", "sex", "dob", "home_address_id", "mailing_address_id", "phone_number", "personal_email", "work_email", "religion", "ward", "stake", "accessibility_accommodations", "high_school", "high_school_gpa", "act_score", "sat_score"])
Student_df.to_csv("Student.csv", index=False)

HomeAddress_df = pd.DataFrame(homeAddresses, columns=["home_address_id", "street_address", "city", "state_province", "country", "zip_code"])
HomeAddress_df.to_csv("HomeAddress.csv", index=False)

MailingAddress_df = pd.DataFrame(mailingAddresses, columns=["mailing_address_id", "street_address", "city", "state_province", "country", "zip_code", "po_box"])
MailingAddress_df.to_csv("MailingAddress.csv", index=False)

Courses_df = pd.DataFrame(courses, columns=["course_id", "semester", "course_name", "course_title", "instructor_id", "meeting_days", "meeting_time", "online_percentage", "min_credits"])
Courses_df.to_csv("Courses.csv", index=False)

Instructor_df = pd.DataFrame(instructors, columns=["instructor_id", "first_name", "last_name"])
Instructor_df.to_csv("Instructor.csv", index=False)

Enrollments_df = pd.DataFrame(enrollments, columns=["enrollment_id", "student_id", "course_id", "credit_hours", "enrollment_type"])
Enrollments_df.to_csv("Enrollments.csv", index=False)

AddDropWithdrawal_df = pd.DataFrame(addDropsWithdrawals, columns=["id", "enrollment_id", "date", "done_by student", "action_type"])
AddDropWithdrawal_df.to_csv("AddDropWithdrawal.csv", index=False)

Grades_df = pd.DataFrame(grades, columns=["grade_id", "enrollment_id", "grade", "gpa_equivalent", "counts_towards_cum_credits", "passing"])
Grades_df.to_csv("Grades.csv", index=False)
```
